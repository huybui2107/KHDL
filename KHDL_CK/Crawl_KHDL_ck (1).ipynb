#**Crawl Data**

##**donghohaitrieu.com**

```python
import bs4
import pandas as pd
# thư viện dể call api
import requests
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

apiURL ="https://donghohaitrieu.com/danh-muc/dong-ho-nam/page/{}"

def getSoup(url):
        req = requests.get(url,
                           headers=headers)  # URL of the website which you want to scrape
        content = req.content  # Get the content
        soup = bs4.BeautifulSoup(content, 'html.parser')
        # print('soup', soup)
        desc = soup.find('div', {'class':'product-loop products-grid product-count-3'})
        
        links = desc.find_all('a', {'class':'product-content-image imageSlider'})
        # print(links)
        # dem = 0
        for link in links :
          allLinks.append(link.get('href'))
```

```python

# Chứa tất cả link sản phẩm
allLinks=[]
# lấy ra tất cả các link từ trang 0 đến trang 57
for i in range(58):
  getSoup(apiURL.format(i))
```

```python
print("Number of link: ",len(allLinks))
```

```python
def get_attribute(url):
  """Hàm tải về một trang web từ một url và trả về một cây HTML"""
  page = requests.get(url,headers=headers)
  return bs4.BeautifulSoup(page.text,"html.parser")
```

###**Get attribute**

```python
Price = []
name = []
Gender = []
Face_Diameter = []
Glass_material = []
Wire_material = []
Apparatus = []
Waterproof = []
Origin =[]
agency = []

for (i,link) in enumerate(allLinks):
  #try catch để nếu xảy ra lỗi thì vẫn tiếp tục crawl k bị dừng lại
  try:
    print(i,link)
    soup=get_attribute(link)
    div = soup.find('div' ,{'class':'thong-tin-san-pham'})
    if div : 
      p = soup.find('p',{'class' : 'price'})
      bdi = p.find('bdi')
      price = bdi.text
      price_clean = bs4.BeautifulSoup(price, 'html.parser').text
      Price.append(price_clean)
      detail = div.find_all('p')
      c1=0
      c2=0
      c3=0
      c4=0
      c5=0
      c6=0
      c7=0
      c8=0
      c9=0
      for i in detail :
        text_content = i.text
        clean_text = bs4.BeautifulSoup(text_content, 'html.parser').text
        clean_text_lower= clean_text.lower()
        txt = ""
        if len(clean_text.split(':')[1]) > 1:
          txt = clean_text.split(':')[1].replace('\n', '')
        else : txt = "NULL"
        if "thương hiệu:" in clean_text_lower :
          agency.append(txt)
          c1 = 1
        elif "xuất xứ" in clean_text_lower :
          Origin.append(txt)
          c2 =1
        elif "chống nước" in clean_text_lower :
          Waterproof.append(txt)
          c3 =1
        elif "máy:" in clean_text_lower :
          Apparatus.append(txt)
          c4=1
        elif "dây đeo:" in clean_text_lower:
          Wire_material.append(txt)
          c5=1
        elif "kính:" in clean_text_lower and "chân kính" not in clean_text_lower :
          Glass_material.append(txt)
          c6=1
        elif "số hiệu sản phẩm" in clean_text_lower :
          name.append(txt)
          c7=1
        elif "đường kính mặt số" in clean_text_lower :
          Face_Diameter.append(txt)
          c8=1
        elif "giới tính"  in clean_text_lower :
          Gender.append(txt)
          c9=1
      if c1 == 0 :
        agency.append("NULL")
      if c2 == 0 :
        Origin.append("NULL")
      if c3 == 0 :
        Waterproof.append("NULL")
      if c4 == 0 :
        Apparatus.append("NULL")
      if c5 == 0 :
        Wire_material.append("NULL")
      if c6 == 0 :
        Glass_material.append("NULL")
      if c7 == 0 :
        name.append("NULL")
      if c8 == 0 :
        Face_Diameter.append("NULL")
      if c9 == 0 :
        Gender.append("NULL")
  except Exception as e:
    print('error: ',i, e)
    continue
```

```python

print(len(Price))
print(len(name))
print(len(Gender ))
print(len(Face_Diameter ))
print(len(Glass_material ))
print(len(Wire_material ))
print(len(Apparatus ))
print(len(Waterproof ))
print(len(Origin ))
print(len(agency ))
```

```python

a = {'name' :name ,'Gender':Gender,'Face_Diameter':Face_Diameter,'Glass_material':Glass_material,'Wire_material':Wire_material,'Apparatus':Apparatus,'Waterproof':Waterproof,'Origin':Origin,'agency':agency , 'Price' :Price}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df.head()
```

###**Save CSV**

```python
dataset = pd.DataFrame(data=df)
dataset.to_csv('dongho_haitrieu.csv', encoding='utf-8-sig', index=False)
```

##**dangquangwatch.vn**

```python
import bs4
import pandas as pd
import requests
headers={'User-Agent':'PostmanRuntime/7.29.0'}

apiURL_2 ="https://www.dangquangwatch.vn/sp/dong-ho-nam.html?&page="

def get_Soup(url):
        req = requests.get(url,
                           headers=headers)  # URL of the website which you want to scrape
        content = req.content  # Get the content
        soup = bs4.BeautifulSoup(content, 'html.parser')
        desc = soup.find('div', {'class':'list-products'})
        links = desc.find_all('a', {'class':'p_info'})
        for link in links :
          all_Links.append(link.get('href'))
```

```python
all_Links=[]
# lấy ra tất cả các link từ trang 1 đến trang 69
for i in range(1,70):
  get_Soup(apiURL_2 +str(i))
```

```python
print("Number of link: ",len(all_Links))
```

```python

hang = ["Epos Swiss" , "Philippe Auguste" , "Jacques lemans" ,"Aries Gold" ,"Atlantic Swiss" ,"Casio" ,"Q&Q" ,"Bruno Sohnle" ,"Stuhrling Original" ,"Citizen"]
name = []
price = []
attr = []
agency = []
mainURL="https://www.dangquangwatch.vn"
for (i,link) in enumerate(all_Links):
  #try catch để nếu xảy ra lỗi thì vẫn tiếp tục crawl k bị dừng lại
  try:
    print(i,mainURL+link)
    soup=get_attribute(mainURL+link)
    div = soup.find('div' ,{'class':'detail_top_right_left'})
    strong = div.find('strong').get_text().lower().replace('đồng hồ' , '').replace('limited','')
    check_ =0 
    # lấy ra tên sản phẩm
    if "tourbillon" in strong :
      agency.append("Tourbillon Memorigin")
      arr = strong.split()
      name.append(arr[len(arr)-1].upper())
      check_ = 1 
    else :
      for check in hang :
        if check.lower() in strong :
          agency.append(check)
          name.append(strong.replace(check.lower(),'').upper())
          check_ = 1
    if check_ == 0 :
      agency.append("NULL")
      name.append("NULL")

    # name.append()
    # lấy ra giá sản phẩm
    price.append(soup.find('span' , {'class':'price'}).get_text())
    dem = 0
    p = soup.find_all('p' , {'class' : 'text2'})
    txt =""
    for i in p:
      if dem == 2 :

        txt += i.text.replace('\n', '').replace(' ','') + ";"
      else :
        txt += i.text.replace('\n', '') + ";"
      dem = dem + 1
    attr.append(txt)
    
    # trích suất bảng thông số
  except Exception as e:
    print('error: ',i, e)
    continue
```

```python
print(len(name))
print(len(price))
print(len(attr))
```

```python

Gender = []
Face_Diameter = []
Glass_material = []
Wire_material = []
Apparatus = []
Waterproof = []
Origin =[]

for n in attr :
  txt = n.split(';');
  Gender.append(txt[7])
  Face_Diameter.append(txt[0])
  Glass_material.append(txt[2])
  Wire_material.append(txt[5])
  Apparatus.append(txt[3])
  Waterproof.append(txt[1])
  Origin.append(txt[8])
```

```python
print(len(Gender))
print(len(Face_Diameter ))
print(len(Glass_material ))
print(len(Wire_material ))
print(len(Apparatus ))
print(len(Waterproof ))
print(len(Origin ))
print(len(agency ))
print(len(name  ))
print(len(price ))
```

```python
a = {'name' :name ,'Gender':Gender,'Face_Diameter':Face_Diameter,'Glass_material':Glass_material,'Wire_material':Wire_material,'Apparatus':Apparatus,'Waterproof':Waterproof,'Origin':Origin,'agency':agency , 'Price' :price}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df.head()
```

###**Save CSV**

```python
dataset = pd.DataFrame(data=df)
dataset.to_csv('dongho_dangquang.csv', encoding='utf-8-sig', index=False)
```

#**danawatch.vn**

```python
import bs4
import pandas as pd
# thư viện dể call api
import requests
headers={'User-Agent':'PostmanRuntime/7.29.0'}

apiURL_3 ="https://danawatch.vn/danh-muc/dong-ho-nam/page/{}"

def get_Soup_2(url):
        req = requests.get(url,
                           headers=headers)  # URL of the website which you want to scrape
        content = req.content  # Get the content
        soup = bs4.BeautifulSoup(content, 'html.parser')
        desc = soup.find_all('div', {'class':'image-none'})
        span_tag = soup.find_all('span', class_='gtm4wp_productdata')

# Get the value of the data-gtm4wp_product_price attribute
        for j in span_tag :
          product_price = j['data-gtm4wp_product_price']
          product_agency =j['data-gtm4wp_product_cat']
          agency.append(product_agency.replace('Đồng hồ',''))
          price.append(product_price)
        for i in desc :
          link = i.find('a').get('href')
          all_Links_3.append(link)
          
        
```

```python
all_Links_3=[]
price = []
agency =[]
# lấy ra tất cả các link từ trang 0 đến trang 4
for i in range(5):
  get_Soup_2(apiURL_3.format(i))
```

```python
print("Number of link: ",len(agency))
print("Price of link: ",len(price))
```

```python
attr = []
for (i,link) in enumerate(all_Links_3):
  #try catch để nếu xảy ra lỗi thì vẫn tiếp tục crawl k bị dừng lại
  try:
    print(i,link)
    soup=get_attribute(link)
    txt = ""
    attribute = soup.find_all('tr')
    dem = 0
    for (j,i) in enumerate(attribute) :
      if j < 12 :
        text = i.get_text(strip=True)
        txt += text + ";"
    attr.append(txt)

      
  except Exception as e:
    print('error: ',i, e)
    continue
```

```python

Face_Diameter = []
Glass_material = []
Wire_material = []
Apparatus = []
Waterproof = []
Origin =[]

name =[]

for (j,n) in enumerate(attr) :
  txt = n.split(';');
  if len(txt) > 10 :        
      
        name.append(txt[1].replace("Mã sản phẩm",''))
        Origin.append(txt[2].replace('Xuất xứ',''))
        Apparatus.append(txt[3].replace('Loại máy',''))
        Glass_material.append(txt[4].replace('Kính',''))
        Face_Diameter.append(txt[6].replace('Đường kính',''))
        Wire_material.append(txt[8].replace('Chất liệu dây',''))
        Waterproof.append(txt[9].replace('Độ chịu nước',''))
        
    
  


      

```

```python
a = {'name' :name ,'Gender':Gender,'Face_Diameter':Face_Diameter,'Glass_material':Glass_material,'Wire_material':Wire_material,'Apparatus':Apparatus,'Waterproof':Waterproof,'Origin':Origin,'agency':agency , 'Price' :price}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
df.head()
```
